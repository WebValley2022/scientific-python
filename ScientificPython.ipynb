```python
# how many pokemons there are for each generation?
pokemons.value_counts("Generation")
```

```python
# and how many Legendary pokemons for each generation?
# hint: select only some of the rows

pokemons[pokemons.Legendary == True].value_counts('Generation')
```

```python
pokemons[pokemons['Type 1'] == 'Fire']
```

```python
# what's the average HP for each pokemon type?
pokemons.groupby(["Type 1", "Type 2"])['HP'].mean().sort_values()
```

```python
from matplotlib import pyplot as plt
import seaborn as sns

fig, ax = plt.subplots(figsize=(10, 10))

sns.scatterplot(
    x="HP",
    y="Defense",
    data=pokemons,
    hue="Legendary",
    ax=ax
)
```

```python
pokemons[
    (pokemons.Legendary)
    & (pokemons.HP < 70)
    & (pokemons.Defense < 50)
]
```

```python
columns = [
    # "Type 1", 
    "Total",
    "HP", 
    "Attack",
    "Defense",
    "Sp. Atk",
    "Sp. Def",
    "Speed",
    "Generation"
]
```

```python
sns.pairplot(
    data=pokemons[columns]
)
```

```python
fig, ax = plt.subplots(figsize=(20, 5))
sns.boxplot(
    x="Type 1",
    y="Total",
    data=pokemons,
    ax=ax
)
```
